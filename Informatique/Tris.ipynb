# Tri par sélection

L'idée est simple: étant donné une liste `L` à trier, on détermine l'emplacement du plus petit élément et on l'échange avec le premier élément. On recommence ensuite avec les éléments restants jusqu'à arriver au bout de la liste. La difficulté ici est de déterminer l'emplacement du minimum (ou plutôt de sa première apparition)

__Vous devez savoir donner le code de la fonction suivante!__

```python
def rang_min(L,i):
    '''
    entrée: L liste et i indice
    sortie: rang  de la première apparition de min(L[i,:])
    '''
    rang=i
    while i<len(L):
        if L[i]<L[rang]:
            rang=i
        i+=1
    return rang
```

```python
def tri_sel(L):
    i=0
    while i<len(L):
        r=rang_min(L,i)
        L[i],L[r]=L[r],L[i]
        i+=1
```

```python
L=[2,1,4,1,6]
tri_sel(L)
L
```

```python
def tri_sel_rec(L,i):
    if i<len(L):
        r=rang_min(L,i)
        L[i],L[r]=L[r],L[i]
        tri_sel_rec(L,i+1)
```

```python
L=[2,1,4,1,6]
tri_sel_rec(L,0)
L
```

```python
def tri_sel_rec2(L):
    if len(L)==1:
        return L
    else:
        r=rang_min(L,0)
        L[r],L[0]=L[0],L[r]
        return([L[0]]+tri_sel_rec2(L[1:]))
```

```python
L=[2,1,4,1,6]
print(tri_sel_rec2(L))
```
